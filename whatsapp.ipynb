```python
import pandas as pd
import numpy as np
import re
import emoji
```

```python
f=open('WhatsApp Chat with °•_T8 Class_•° FOREVER(off).txt','r',encoding='utf-8')
data=f.read()
```

```python
print(type(data))
print(data)
```

```python
#lets create a dataframe one with message column and one with data colummns
pattern='\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s(?:am|pm)\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates= re.findall(pattern,data)
dates = [date.replace('\u202f', '') for date in dates]
dates
```

```python
df=pd.DataFrame({'user_message':messages,'date':dates})

df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y, %I:%M%p - ')

df.head()
```

```python
df.shape
```

```python
users=[]
messages=[]
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user']=users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)
df.head()
```

```python
df['day_name']=df['date'].dt.day_name()
```

```python
df['day_name'].value_counts()
```

```python
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour==23:
        period.append(str(hour)+"-"+str('00'))
    elif hour==0:
        period.append(str('00')+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
```

```python
df['period']=period
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name',columns='period',values='message',aggfunc='count').fillna(0))
plt.yticks(rotation="horizontal")
plt.xlabel("Time Period")
plt.ylabel("Day")
plt.show()
```
